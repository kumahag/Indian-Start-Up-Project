# Indian Start-Up Data Analysis Project

## Hypothesis

Null Hypothesis 1 : Start –ups  in Large cities receive more funding  based on business location

 Null Hypothesis 2: Tech Start-up Companies   largely receive higher funding amounts than businesses in other sectors



###Main Objective: 
 Investigate India Start-up Funding Ecosystem  to determine best  funding options, service industry and or location to Start-up a business.
 
###Questions

1.How much funding do Start-up’s generally get in India?
2. How has funding changed with time and what are the sources of these funds (Investors)?
3. Which Sectors/Industries are favored by Investors in terms of funding amount/ volume?
4.In which year was most funding (amount received) and number of Start-up’s funded?
5.Is funding received related to location of the business?
